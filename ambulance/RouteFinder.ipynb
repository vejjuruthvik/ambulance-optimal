```python
import osmnx as ox
import networkx as nx
import folium

ox.config(log_console=True, use_cache=True)
from geopy.geocoders import Nominatim
locator = Nominatim(user_agent = "myapp")

start_latlng = (start_location.latitude, start_location.longitude)
end_latlng = (end_location.latitude, end_location.longitude)
mode      = 'bike'      
optimizer = 'length'

graph = ox.graph_from_point((start_location.latitude, start_location.longitude), network_type = mode)
orig_node = ox.get_nearest_node(graph, start_latlng)
dest_node = ox.get_nearest_node(graph, end_latlng)
```

```python
shortest_route = nx.shortest_path(graph,
                                  orig_node, dest_node,
                                  weight=optimizer)

shortest_route_map = ox.plot_route_folium(graph, shortest_route, tiles='openstreetmap')
shortest_route_map
```

```python
start_latlng = start_latlng
end_latlng   = end_latlng
start_marker = folium.Marker(
            location = start_latlng,
            popup = start_location,
            icon = folium.Icon(color='green'))
end_marker = folium.Marker(
            location = end_latlng,
            popup = end_location,
            icon = folium.Icon(color='red'))
# add the circle marker to the map
start_marker.add_to(shortest_route_map)
end_marker.add_to(shortest_route_map)
shortest_route_map
```
